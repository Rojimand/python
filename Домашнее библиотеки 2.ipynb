# Задание 1

```python
import pandas as pd
data = pd.read_csv('ratings.csv')
data.head()
```

```python
def rating(data):
    if (2.0 >= data['rating']):
        return 'низкий рейтинг'
    elif (2.0 < data['rating']) and (data['rating'] <= 4):
        return 'средний рейтинг'
    else: 
        return 'высокий рейтинг'
    return False
```

```python
data['class'] = data.apply(rating, axis=1)
data.head()
```

# Задание 2

```python
import pandas as pd
keywor = pd.read_csv('keywords.csv')
keywor.head(50)
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def geo_classifier(row):
    for region, cities in geo_data.items():
        for city in cities:
            if city in row['keyword']:
                return region
    return 'undefined'
```

```python
keywor['region'] = keywor.apply(geo_classifier, axis=1)
keywor.head()
```
